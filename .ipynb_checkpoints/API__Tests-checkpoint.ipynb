```python
# Importing Data From Yahoo Finance

import pandas as pd
import yfinance as yf
from pandas.tseries.offsets import BMonthEnd

# Fetch the historical data for AAPL, at a specific frequancy (M)
ticker = '^VIX'
data = yf.download(ticker, start='1990-01-01', end='2022-07-21', interval='1mo')

# Extract the OHLC prices
ohlc = data[['Open', 'High', 'Low', 'Close', 'Volume']]

# Create a pandas DataFrame from the OHLC data
df = pd.DataFrame(ohlc)
df.index = pd.to_datetime(df.index)  # Convert index to datetime

df.index = pd.to_datetime(df.index) + BMonthEnd(0)  # Add BMonthEnd(0) to the index
df.index = df.index.strftime('%m/%d/%Y')

df
```
